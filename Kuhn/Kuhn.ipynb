```python
#solve with bisection

import numpy as np
import copy as cp

def get_EV(x):
    node = [0]*6
    node[0]= -(1-x[0])+x[0]*((1-x[3])-2*x[3])
    node[1] = -(1-x[0])-2*x[0]
    node[2] = (1-x[2])+x[2]*(-(1-x[4])+2*x[4])
    node[3] = -(1-x[4])-2*x[4]
    node[4]= (1-x[1])*((1-x[2])+x[2]*2)+x[1]
    node[5]= (1-x[1])+x[1]*((1-x[3])+x[3]*2)
    return sum(node)
def sweepAll(which,x_in,x_val):
    s=[0]*4
    si= 0 
    x = cp.copy(x_in)
    x[which] = x_val
    tmp0 = cp.copy(x)
    tmp1 = cp.copy(x)        
    for i in range(5):
        if i != which:
            tmp0[i] = 0
            tmp1[i] = 1
            s[si] = abs(get_EV(tmp1)-get_EV(tmp0))
            tmp0[i]=x[i]
            tmp1[i]=x[i]
            si+=1
    return sum(s)
a = 1/2
g = 1/2
c = 1/2
n = 1/2
b = 1/2 

x = [a,g,c,n,b]

for which in range(5):
#     for j in np.linspace(0, 1.0, 101, endpoint=True):
    it=0
    cur_x = 0.5
    l=0
    r=1
    print(which,'------')
    while it<100:
        it=it+1
        mid = (r+l)/2
        y1 = sweepAll(which,x,mid)
        y2 = sweepAll(which,x,min((mid+0.0001,1)))
        dely = y2-y1
        
#         print(round(l,3),round(r,3))
        if dely<0:
            l= mid
        else:
            r = mid
    x[which] = mid
print(x)
```
